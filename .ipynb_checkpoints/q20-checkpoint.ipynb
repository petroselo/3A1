```python
from __future__ import print_function
from ipywidgets import interact, interactive, fixed, interact_manual
import ipywidgets as widget

import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

```python
%matplotlib inline

a=2
d=4
G=10
@interact(G=1.0)
def f(G):
    xvect = np.linspace(-3,10, 501)
    yvect = np.linspace(-5,5,501)
    x, y = np.meshgrid(xvect, yvect, indexing='ij')
    z = x + 1.j*y
    F = -1.j*G*np.log(z) +1.j*G*np.log(z-a**2/d) -1.j*G*np.log(z-d)

    plt.contour(x, y, np.imag(F), np.linspace(-100, 100, 202))
    plt.gca().set_aspect("equal")
    plt.show()
```
