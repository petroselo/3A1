```python
#imports
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
%matplotlib inline

A = 0.5 #m/2piUh
h = 1
xvect = np.linspace(-5, 5, 1001)
yvect = np.linspace(0, 5, 1001)
x, y = np.zeros((xvect.size, yvect.size)), np.zeros((xvect.size, yvect.size))
for i in range(xvect.size):
    for j in range(yvect.size):
        x[i,j], y[i,j] = xvect[i], yvect[j]
z = x + 1.j*y #another big mesh
F = z - A*(np.log(z-1.j) + np.log(z+1.j) + 2*np.log(h))

ψ = np.imag(F)

plt.contour(x, y, ψ, np.linspace(-2*np.pi, 4*np.pi,31 ), colors="blue")
plt.gca().set_aspect("equal")
plt.show()

#fig = plt.figure()
#ax = fig.add_subplot(111, projection='3d')
#ax.plot_surface(x, y, ψ)
#plt.show()



```
