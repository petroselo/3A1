```python
#imports
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

```python
phivec = np.linspace(-3., 24, 51)
psivec = np.linspace(-np.pi, np.pi, 21)
phi, psi = np.meshgrid(phivec, psivec, indexing='ij') # columns are constant psi
F = phi + 1.j*psi
z = F-np.exp(-F)

plt.plot(np.real(z), np.imag(z))
plt.show()
```
