```python
#q10
e=0.00001
steps = 1000
c=1
s= np.linspace(e-3*c,3*c-e,steps)
h=1
U=5
a=0.09
rho = 1.2

def circulations_and_lifts(c, s, h, U, a):
    #after applying collocation point conditions
    x1 = s+c/2
    x2 = s-c/2
    r12 = x1*x1 + h*h
    r22 = x2*x2 + h*h
    g1= U*a*(-1 + 2*r12/(c*x1)) * 1/((x2)/(2*np.pi*r22) + (2*np.pi*r12)/(x1*(np.pi*c)**2))
    g2 = U*a*(1+c*x2/(2*r22)) * 1/(1/(np.pi*c) + c*x1/(2*r12))
    
    rg = np.sqrt(h*h+s*s)
    u1 = g2/(2*np.pi*rg**2) * h + U*np.cos(a)
    v1 = g2/(2*np.pi*rg**2) * s + U*np.sin(a)
    u2 = g1/(2*np.pi*rg**2) * (-h) + U*np.cos(a)
    v2 = g1/(2*np.pi*rg**2) * (-s) + U*np.sin(a)
    
    l1 = rho*g1*u1[0]
    l2 = rho*g2*u2[0]
    
    return g1, g2, l1, l2

g1, g2, l1, l2 = circulations_and_lifts(c,s,h,U,a)
g1, g2 = g1/(np.pi*U*a*c), g2/(np.pi*U*a*c)
l1, l2= l1/(rho*np.pi*U*U*a*c), l2/(rho*np.pi*U*U*a*c)


plt.plot(s, g1, label='G1')
plt.plot(s, g2, label='G2')
plt.plot(s, g1+g2, label='G1 + G2')
plt.legend()
plt.show()

plt.plot(s, l1, label='L1')
plt.plot(s, l2, label='L2')
plt.plot(s, l1+l2, label='L1 + L2')
plt.legend()
plt.show()


```
