```python
#5
%matplotlib inline

import matplotlib
import numpy as np
import matplotlib.pyplot as plt

θ = np.linspace(0., 2. * np.pi, 101)
z = -1 * np.exp(1.j * θ)
ζ = z + 1. / z
plt.plot(np.real(ζ), np.imag(ζ))
plt.axis('equal')
plt.show()
```

```python
#5d
@interact(rmax=1.0, theta=(0.0,2*np.pi))
def f(rmax, theta):
    r = np.linspace(0.001, rmax + 0.0, 101)
    z = r * np.exp(1.j*theta)
    ζ = z + 1. / z
    plt.plot(np.real(ζ), np.imag(ζ))
    plt.axis( xmin=-1000, xmax=1000, ymin=-1000, ymax=1000)
    plt.show()
```

```python
#6
import numpy as np
import matplotlib.pyplot as plt
h0, a, x0, b, y0 = 10. , 0.25, 2., 4., -1.
x = np.linspace(-10., 15., 26)
y = np.linspace(-5., 5., 11)
xmat, ymat = np.zeros((x.size, y.size)), np.zeros((x.size, y.size))
for i in range(x.size):
    for j in range(y.size):
        xmat[i, j], ymat[i,j] = x[i], y[j]
h = h0 - a*(xmat-x0)**2 - b*(ymat-y0)**2

plt.plot(x, h[:, 6]); plt.show() #6 is the middle of the array so shows at a fixed value of y
plt.plot(y, h[11, :]); plt.show()

plt.contour(xmat, ymat, h, [-50, -10, -1, 0, 2, 7]); plt.show()
```

```python
#8
import numpy as np
import matplotlib.pyplot as plt
r = np.linspace(0., 5., 51)
theta = np.linspace(0.,2.*np.pi,101)
xmat, ymat, psi = np.zeros((r.size, theta.size)), np.zeros((r.size, theta.size)), np.zeros((r.size, theta.size)) 
for i in range(r.size):
    for j in range(theta.size):
        xmat[i, j] = r[i] * np.cos(theta[j])
        ymat[i, j] = r[i] * np.sin(theta[j])
        psi[i, j] = r[i] * np.exp(-r[i]) * (np.cos(theta[j]))**2
plt.contour(xmat, ymat, psi)
plt.show()
```

```python
#8d
import numpy as np
import matplotlib.pyplot as plt
x, y = np.linspace(-5, 5, 51), np.linspace(-5, 5, 51)
xmat, ymat, rmat, thmat, psi = np.zeros((x.size, y.size)), np.zeros((x.size, y.size)), np.zeros((x.size, y.size)), np.zeros((x.size, y.size)), np.zeros((x.size, y.size))
for i in range(x.size):
    for j in range(y.size):
        xmat[i,j] = x[i]
        ymat[i,j] = y[j]
rmat = (xmat**2 + ymat**2)**0.5
thmat = np.arctan2(ymat, xmat)
psi = rmat*np.exp(-rmat)*(np.cos(thmat))**2
plt.contour(xmat, ymat, psi)
plt.show()
```

```python
#9
import numpy as np
import matplotlib.pyplot as plt
x, y = np.linspace(-10, 10, 51), np.linspace(-10, 10, 51)
cmat = np.zeros((x.size, y.size), dtype=complex)
for i in range(x.size):
    for k in range(y.size):
        cmat[i, k] = x[i] + x[k]*1j
print(cmat)
plt.contour(np.real(cmat), np.imag(cmat), np.real(cmat**2), colors='red') # phi or the real part
plt.contour(np.real(cmat), np.imag(cmat), np.imag(cmat**2), colors='blue') # psi or the imag part
plt.show()
```
